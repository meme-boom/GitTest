```python
list1 = list([['한빛', '남자', '20', '180'], ['한결', '남자', '21', '177'], ['한라', '여자', '20', '160']])
col_names = ['이름', '성별', '나이', '키']
df = pd.DataFrame(list1, index=['1반','2반','3반'],columns=col_names)
df.index
df.values
df.value_counts()
df['나이'][1]
```
